```python
import spotipy
import json
import pandas as pd
from spotipy.oauth2 import SpotifyOAuth
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
credentials = json.load(open('authorization.json'))
client_id = credentials['client_id']
client_secret = credentials['client_secret']
client_credentials_manager = SpotifyClientCredentials(client_id=client_id,client_secret=client_secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
sp
```

```python
new_tracks = sp.new_releases()
```

```python
# album_type, artists, available_markets, external_urls, href, id, images, name, 
# release_date, release_date_precision, total_tracks, type, uri
```

```python
album_type = []
artists = [] # further division
available_markets = []
external_urls = []
href = []
id = []
images = [] # further division
name = []
release_date = []
release_date_precision = []
total_tracks = []
type = []
uri = []
```

```python
artists_id = []
artists_name = []
```

```python
count = 0
for i, element in enumerate(new_tracks["albums"]["items"]):
    album_type.append(element["album_type"])
    artists.append(element["artists"])
    available_markets.append(element["available_markets"])
    external_urls.append(element["external_urls"])
    href.append(element["href"])
    id.append(element["id"])
    images.append(element["images"])
    name.append(element["name"])
    release_date.append(element["release_date"])
    release_date_precision.append(element["release_date_precision"])
    total_tracks.append(element["total_tracks"])
    type.append(element["type"])
    uri.append(element["uri"])
    temp_id = []
    temp_name = []
    for j, artist in enumerate(element["artists"]):
        temp_id.append(artist["id"])
        temp_name.append(artist["name"])
    artists_id.append(temp_id)
    artists_name.append(temp_name)
```
