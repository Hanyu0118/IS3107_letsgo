# Remove duplicates + One-hot-encode tracks total 1-4

```python
import pandas as pd 
df = pd.read_csv('../Track_Genre/tracks_total_1.csv')
df2 = pd.read_csv('../Track_Genre/tracks_total_2.csv')
df3 = pd.read_csv('../Track_Genre/tracks_total_3.csv')
df4 = pd.read_csv('../Track_Genre/tracks_total_4.csv')
frames = [df,df2,df3,df4]
result = pd.concat(frames)
```

```python
result = result.drop('playlist_id',axis=1)
result = result.drop_duplicates()
```

```python
dummy_df = pd.get_dummies(result['Genre_name'], prefix='', prefix_sep='')
```

```python
one = pd.concat([result['track_id'], dummy_df], axis=1)
```

# Load the table

```python
schema = [
    bigquery.SchemaField("track_id","STRING"),
    bigquery.SchemaField("Alternative","STRING"),
    bigquery.SchemaField("Blues","STRING"),
    bigquery.SchemaField("Cantopop","STRING"),
    bigquery.SchemaField("Classical","STRING"),
    bigquery.SchemaField("Dance Electronic","STRING"),
    bigquery.SchemaField("Folk & Acoustic","STRING"),
    bigquery.SchemaField("Hip-Hop","STRING"),
    bigquery.SchemaField("Indie","STRING"),
    bigquery.SchemaField("Instrumental","STRING"),
    bigquery.SchemaField("Jazz","STRING"),
    bigquery.SchemaField("K-Pop","STRING"),
    bigquery.SchemaField("Latin","STRING"),
    bigquery.SchemaField("Malay Pop","STRING"),
    bigquery.SchemaField("Mandopop","STRING"),
    bigquery.SchemaField("Metal","STRING"),
    bigquery.SchemaField("Musik Indonesia","STRING"),
    bigquery.SchemaField("Pop","STRING"),
    bigquery.SchemaField("R&B","STRING"),
    bigquery.SchemaField("Rock","STRING"),
]
```

```python
table_id = "snappy-boulder-378707.TrackGenre.Trackgenre"
table = bigquery.Table(table_id, schema=schema)
table = client.create_table(table, timeout=30)  # Make an API request.
print("Created table {}.{}.{}".format(table.project, table.dataset_id, table.table_id))
```

```python
job_config = bigquery.LoadJobConfig(
    schema = schema,
    source_format=bigquery.SourceFormat.CSV,
    skip_leading_rows=1,
    autodetect=True,
)

file_path = 'tracks_genre_1.csv'
table_id = "snappy-boulder-378707.TrackGenre.Trackgenre"

with open(file_path, "rb") as source_file:
    job = client.load_table_from_file(source_file, table_id, job_config=job_config)

job.result()  # Waits for the job to complete.

table = client.get_table(table_id)  # Make an API request.
print(
    "Loaded {} rows and {} columns to {}".format(
        table.num_rows, len(table.schema), table_id
     )
)
```
