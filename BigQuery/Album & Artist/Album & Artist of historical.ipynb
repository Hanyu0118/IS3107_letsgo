```python
import ast
def compute_markets(s):
    markets = s.replace('[','').replace(']','').replace("'",'').replace(' ', '').split(',')
    if markets[0] == '':
        return 0
    else:
        return len(markets)

```

```python
from datetime import date
album_info['id'] = data['album'].apply(lambda x: ast.literal_eval(x)['id'])
album_info['name'] = data['album'].apply(lambda x: ast.literal_eval(x)['name'])
album_info['total_tracks'] = data['album'].apply(lambda x: ast.literal_eval(x)['total_tracks'])
album_info['release_date'] = data['album'].apply(lambda x: ast.literal_eval(x)['release_date'])
album_info['extract_date'] = date.today()
album_info['available_markets'] = data['available_markets'].apply(compute_markets)
```

```python
album_info.head()
```

```python
album_info = album_info[album_info.release_date != '0000']
```

```python
album_info['release_date'] = album_info['release_date'].apply(lambda x: pd.to_datetime(x).date())
```
